```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

```python
#get the total inique players by their screen names
totalplayers=purchase_data["SN"].unique().size
#assing dataframe
totalplayers_df=pd.DataFrame([[totalplayers]])
totalplayers_df.colums=[['Total Players']]
totalplayers_df.head()

```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#obtaion total number of unique items
unique_items=purchase_data["Item ID"].unique().size

#cal avg price
avg_price=round(purchase_data["Price"].mean(),2)
#format to $0.00
#avg_price='$'+ avg_price.astype(str)

#get the total purchases
number_purchase=purchase_data["Purchase ID"].size

#total revenue
total_revenue=round(purchase_data["Price"].sum(),2)
#fromat rev to $0.00
total_revenue='$'+total_revenue.astype(str)

#write new dataframe
summary_df=pd.DataFrame([[unique_items,avg_price,number_purchase,total_revenue]])
summary_df.columns=['Number of Unique Items','Average Price','Number of Purchases','Total Revenue']

summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#new dataframe for summary
summary_df2=pd.DataFrame(columns=['Percentage of Players','Total Count'], index =['Male','Female','Other/NonDisclosed'])
#get size of the inique purchase ID rows for each gender
male_df=purchase_data.loc[purchase_data['Gender']=="Male",["Gender","Purchase ID","Price"]]
male_rows=male_df["Purchase ID"].unique().size
female_df=purchase_data.loc[purchase_data['Gender'] == "Female",["Gender","Purchase ID","Price"]]
female_rows=female_df["Purchase ID"].unique().size
other_df=purchase_data.loc[(purchase_data['Gender'] != "Male") & (purchase_data['Gender'] != "Female"),["Gender","Purchase ID","Price"]]
other_rows=other_df["Purchase ID"].unique().size

#get the percentages
male_percent=round(male_rows/totalplayers*100,2)
female_percent=round(female_rows/totalplayers*100,2)
other_percent=round(other_rows/totalplayers*100,2)
#write the series to the new dataframe
summary_df2.loc['Male'] = pd.Series({'Percentage of Players':male_percent,'Total Count':male_rows})
summary_df2.loc['Female'] = pd.Series({'Percentage of Players':female_percent,'Total Count':female_rows})
summary_df2.loc['Other/NonDisclosed'] = pd.Series({'Percentage of Players':other_percent,'Total Count':other_rows})
summary_df2.head()








```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = ([0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999])
group_names = (["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"])

#bin the data based on age and calculate the percentages and write to a new dataframe
bin_df=pd.DataFrame(round((((pd.cut(purchase_data["Age"], age_bins, labels=group_names)).value_counts())/totalplayers*100),2))

#rename the new dataframe column to "percentage of players"
bin_df.rename(columns={'Age': 'Percentage of Players'},inplace = True)

#bin the data based on age and count the total count for each age group
bin_df["Total Count"]=pd.cut(purchase_data["Age"], age_bins, labels=group_names).value_counts()
#reindex the dataframe to follow the order of the group names as in the list "group_names"
bin_df=bin_df.reindex(group_names)
bin_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#get the item Id counts from purchase data for each screen name
top_spenders_df = purchase_data['Item ID'].groupby(purchase_data['SN']).count()

# Create the initial column in the Dataframe
top_spenders_df = pd.DataFrame(data=top_spenders_df)
top_spenders_df.columns = ['Purchase Count']

# Calculate the column values
#average purchase price for each screen name
top_spenders_df['Average Purchase Price'] = round(purchase_data['Price'].groupby(purchase_data['SN']).mean(),2)
#total purchase price for each screen name
top_spenders_df['Total Purchase Value'] = purchase_data['Price'].groupby(purchase_data['SN']).sum()

#sort the dataframe by total purchase values in descending order
top_spenders_df.sort_values(by=['Total Purchase Value'], ascending=False, inplace=True)

# Formatting the average price and total price in the format $0.00
top_spenders_df['Average Purchase Price'] = top_spenders_df['Average Purchase Price'].map('${:,.2f}'.format)
top_spenders_df['Total Purchase Value'] = top_spenders_df['Total Purchase Value'].map('${:,.2f}'.format)

#output for top 5 spenders
top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Get the Purchase Count and Total Purchase Value from purchase_data
#get the item id,item name,price and group by price and athen ggregate the data to get the total count for each group and the total sum for each group
most_items_df = purchase_data.groupby(['Item ID', 'Item Name','Price'])['Price'].agg(['count','sum'])
#write the result to two columns
most_items_df.columns = ['Purchase Count', 'Total Purchase Value']

# Reformatting the index 
most_items_df.reset_index(inplace=True)
most_items_df.set_index(['Item ID','Item Name'] ,inplace=True)

# Renaming Dataframe columns
most_items_df = most_items_df[['Purchase Count', 'Price', 'Total Purchase Value']]

# Sorting by most popular in descending order (Highest Purchase Count=most popular)
most_popular_items_df = most_items_df.sort_values(by='Purchase Count', ascending=False)

# Format the price and total purchase value in the format $0.00
most_popular_items_df['Price'] = most_popular_items_df['Price'].map('${:,.2f}'.format)
most_popular_items_df['Total Purchase Value'] = most_popular_items_df['Total Purchase Value'].map('${:,.2f}'.format)

#output the 5 most popular
most_popular_items_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Sort by most profitable (highest Total Purchase Value=most profitable) in descending order
most_profitable_items_df = most_items_df.sort_values(by='Total Purchase Value', ascending=False)

# Format into $0.00 for price and total purchase value
most_profitable_items_df['Price'] = most_profitable_items_df['Price'].map('${:,.2f}'.format)
most_profitable_items_df['Total Purchase Value'] = most_profitable_items_df['Total Purchase Value'].map('${:,.2f}'.format)

#output the 5 most profitable games
most_profitable_items_df.head()
```
